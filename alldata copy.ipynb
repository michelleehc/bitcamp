```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.concat(map(pd.read_csv, ['atlanta.csv', 'fargo.csv', 'neworleans.csv', 'okcity.csv', 'omaha.csv']), ignore_index=True)
title = "Other States"
print(df)
```

```python
df = pd.concat(map(pd.read_csv, ['tampa.csv', 'tallahassee.csv', 'miami.csv']), ignore_index=True)
title = "Florida"
print(df)
```

```python
data = df
years = np.array(df['PayYear'].tolist())
months = np.array(df['PayMonth'].tolist())
years = years.astype(float) 
months = months.astype(float)
months = months * 1/12 
x = np.add(years, months)
y = df['Original_Interest_Rate'].tolist()
heatmap, xedges, yedges = np.histogram2d(x, y, bins=100)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
fig, ax = plt.subplots(figsize = (12, 8))
ax.legend()
plt.legend(bbox_to_anchor=(1.04, 0.5), loc='center left', borderaxespad=0, title="Class")
plt.clf()
plt.imshow(heatmap.T, cmap= 'inferno', extent=extent, origin='lower', norm= 'symlog')
plt.xlabel("Year")
plt.ylabel("Interest rate (%)")
plt.title(title)
plt.show()
```
