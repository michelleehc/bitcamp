```python
from collections import Counter

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.concat(map(pd.read_csv, ['atlanta.csv', 'fargo.csv', 'neworleans.csv', 'okcity.csv', 'omaha.csv']), ignore_index=True)
title = "Other States"
print(df)
```

```python
df = pd.concat(map(pd.read_csv, ['tampa.csv', 'tallahassee.csv', 'miami.csv']), ignore_index=True)
title = "Florida"
print(df)
```

```python
data = df
years = np.array(df['PayYear'].tolist())
months = np.array(df['PayMonth'].tolist())
years = years.astype(float) 
months = months.astype(float)
months = months * 1/12 
x = np.add(years, months)
y = df['Original_Interest_Rate'].tolist()
heatmap, xedges, yedges = np.histogram2d(x, y, bins=100)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
fig, ax = plt.subplots(figsize = (12, 8))
ax.legend()
plt.legend(bbox_to_anchor=(1.04, 0.5), loc='center left', borderaxespad=0, title="Class")
plt.clf()
plt.imshow(heatmap.T, cmap= 'inferno', extent=extent, origin='lower', norm= 'symlog')
plt.xlabel("Year")
plt.ylabel("Interest rate (%)")
plt.title(title)
plt.show()
```

```python
data = df
years = np.array(df['PayYear'].tolist())
months = np.array(df['PayMonth'].tolist())
years = years.astype(float) 
months = months.astype(float)
months = months * 1/12 
x = np.add(years, months)
y = np.array(df['Foreclosure_date'].tolist())
y = np.nan_to_num(y)
y = y.astype(bool)
heatmap, xedges, yedges = np.histogram2d(x, y, bins=200)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
fig, ax = plt.subplots(figsize = (12, 8))
ax.legend()
plt.legend(bbox_to_anchor=(1.04, 0.5), loc='center left', borderaxespad=0, title="Class")
plt.clf()
plt.imshow(heatmap.T, cmap= 'inferno', extent=extent, origin='lower', norm= 'symlog', aspect= 10)
plt.xlabel("Year")
plt.ylabel("LTV")
plt.title(title)
plt.show()
```

```python
data = df
y = np.array(df['Foreclosure_date'].tolist())
y = [x for x in y if str(x) != 'nan']
y = np.array(y)
y = y % 10000
y = y.astype(int)

sorted_list = sorted(y)
sorted_counted = Counter(sorted_list)

range_length = list(range(max(y))) # Get the largest value to get the range.
data_series = {}

for i in range_length:
    data_series[i] = 0 # Initialize series so that we have a template and we just have to fill in the values.

for key, value in sorted_counted.items():
    data_series[key] = value

data_series = pd.Series(data_series)
x_values = data_series.index

# you can customize the limits of the x-axis
# plt.xlim(0, max(some_list))
plt.bar(x_values, data_series.values)
plt.xlim(2000, 2024)
plt.xlabel("Year")
plt.ylabel("Forclosures")
plt.title(title)
plt.show()
```
